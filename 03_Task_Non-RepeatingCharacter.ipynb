```python
from collections import Counter      # Import Counter to count occurrences

def first_unique_char(s):            # Normal function definition
    char_count = Counter(s)          # Count occurrences of each character

    for char in s:                   # Iterate through the string in order
        if char_count[char] == 1:    # Find the first character with count 1
            return char

    return "None"                    # If no unique character is found
```

```python
s = "swiss"
print(first_unique_char(s))
```

```python
# Time Complexity = O(n)
# Space Complexity = O(n) (if all characters are unique, storing counts for all).
#                  = O(1) (if the string is empty or very short).
```
