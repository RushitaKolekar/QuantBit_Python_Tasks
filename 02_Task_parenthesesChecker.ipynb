```python
def is_valid(s):                                     #normal function defination
  stack = []                                         #keep track of open brackets
  mapping = { ')': '(', '}': '{', ']': '[' }         #matching pairs
  for char in s:
    if char in mapping:                              #if it's closing bracket
      top_element = stack.pop() if stack else '#'    #pop or use a dummy
      if mapping[char] != top_element:               #check if it matches
        return False
    else:                                            #it's an opening bracket
      stack.append(char)
  return not stack                                   # return true if stack empty(balanced), otherwise false
```

```python
s1 = "{[()]}"
print(is_valid(s1))  # Output: True

s2 = "{[(])}"
print(is_valid(s2))  # Output: False
```

```python
# Time Complexity = O(n)
# Space Complexity = O(n) - when all characters are opening brackets
#                  = O(1) - if the input is empty or instantly balanced
```
