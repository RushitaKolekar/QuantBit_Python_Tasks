```python
from collections import defaultdict             # Import defaultdict to group words

def group_anagrams(words):
    anagrams = defaultdict(list)                # Dictionary to store grouped anagrams

    for word in words:
        sorted_word = "".join(sorted(word))      # Sort the word (e.g., "eat" â†’ "aet")
        anagrams[sorted_word].append(word)       # Add the word to the corresponding group

    return list(anagrams.values())               # Convert dictionary values to list
```

```python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams(words))
```

```python
#Time Complexity = O(n * k log k)
# n = Number of words
# k log k = Sorting each word (where k is the word length)

# Space Complexity = O(n * k)  Storing all words in the dictionary.
```
