```python
def longest_consecutive(nums):                   # Normal function definition
    if not nums:                                 # Handle empty list case
        return 0

    num_set = set(nums)                          # Convert list to set for O(1) lookups
    longest = 0                                  # Store the longest sequence length

    for num in num_set:
        if num - 1 not in num_set:               # Start of a new sequence
            current_num = num
            current_streak = 1

            while current_num + 1 in num_set:    # Continue the sequence
                current_num += 1
                current_streak += 1

            longest = max(longest, current_streak)  # Update longest

    return longest                                 # Return the maximum sequence length
```

```python
nums = [100, 4, 200, 1, 3, 2]
print(longest_consecutive(nums))
```

```python
# Time Complexity = O(n)
# Space Complexity = O(n)
```
