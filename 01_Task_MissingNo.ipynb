```python
def find_missing_no(arr:list) -> int:
  n= len(arr)                     #give size of list
  total_sum = n*(n+1)//2          #sum of no from 0 to n
  actual_sum = sum(arr)           #sum of given no
  return total_sum - actual_sum   # return missing no
```

```python
arr=[3,0,1]
print(find_missing_no(arr))
```

```python
# Time complexity = O(n),
# Space Complexity = O(1)
```
