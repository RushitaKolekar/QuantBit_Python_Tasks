```python
def find_kth_largest(nums: list, k: int) -> int:
    nums.sort(reverse=True)                      # Sort the list in descending order
    return nums[k - 1]                           # Return the k-th largest element(k-1 index)
```

```python
nums = [3, 2, 1, 5, 6, 4]
k = 2
print(find_kth_largest(nums, k))
```

```python
# Time Complexity = O(n log n)
# Space Complexity = O(1)
```
